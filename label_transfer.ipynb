```r
file_path <- "/QRISdata/Q1040/IntermediateData/Quan_Visium2/Visium_LP2_map_C1/outs/"
# spatial_file <- paste(file_path,"filtered_feature_bc_matrix")
# print(spatial_file)
data <- Load10X_Spatial(data.dir = file_path, filename = 'filtered_feature_bc_matrix.h5',
                       assay='Spatial', slice='slice1')
```

```r
data <- SCTransform(data, assay = "Spatial", verbose = FALSE)
```

```r
data <- ScaleData(data, assay='SCT')
data <- RunPCA(data, assay = "SCT", verbose = FALSE)
```

```r
sc_data_dir <- '/QRISdata/Q1851/Minh/skin_sc_ref'
sc_data <- Read10X(data.dir=sc_data_dir)
```

```r
object <- CreateSeuratObject(counts = sc_data, project = "bbc.skin")
object <- SCTransform(object, verbose = FALSE) %>% RunPCA(npcs = 50, verbose = FALSE) %>% 
  RunUMAP(dims = 1:50, umap.method = "umap-learn", metric = "correlation", n.components = 2L)
```

```r
object <- FindNeighbors(object, reduction = "pca", dims = 1:20)
object <- FindClusters(object, resolution = 0.4)
```

```r
#latest dim plot
p1 <- DimPlot(object, reduction = "umap")
p1
```

```r
# old dim plot
p1 <- DimPlot(object, reduction = "umap")
p1
```

```r
markers <- FindAllMarkers(object, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)
markers %>% group_by(cluster) %>% top_n(n = 20, wt = avg_logFC)
ccc
```

```r
df_object <- data.frame(sapply(markers,c))
df_object
```

```r
write.csv(df_object, 'gene_marker_skin.csv')
```

```r
new.cluster.ids <- c("macrophages/dendritic", "fibroblasts", "fibroblasts", 'fibroblasts',
                     "vascular endothelial", "epidermal stem cell", "differentiated keratinocytes", 'T cells',
                    'pericytes', 'fibroblasts', 'erythrocytes', 'lymphatic endothelial cells', 'melanocytes',
                    'junk 1', 'junk 2')
names(new.cluster.ids) <- levels(object)
object <- RenameIdents(object, new.cluster.ids)
DimPlot(object, reduction = "umap", label = TRUE, pt.size = 0.5) + NoLegend()
```

```r
new_clusters <- subset(object, idents = c('macrophages/dendritic', 'epidermal stem cell','differentiated keratinocytes', 'T cells'))
```

```r
new_clusters <- ScaleData(new_clusters)
new_clusters <- RunPCA(new_clusters)
```

```r
saveRDS(new_clusters,'transfer_reference_4clusters.rds')
```

```r
# subset_interest2 <- readRDS('transfer_reference_without_fibroblasts.rds')
pdf('Umap_selected_cluster.pdf')
DimPlot(new_clusters, reduction = "umap", label = TRUE, pt.size = 0.5) + NoLegend()
dev.off()
```

```r
new_clusters[['cell_type']] <- Idents(new_clusters)
levels(new_clusters)
```

```r
anchors <- FindTransferAnchors(reference = new_clusters, query = data, query.assay='SCT',normalization.method = "SCT", npcs = 50,
                               dims = 1:50, k.anchor = 5, k.filter = 200)
predict_assay_subset <- TransferData(anchorset = anchors, refdata = new_clusters$cell_type, prediction.assay = TRUE, 
    weight.reduction = data[["pca"]], dims = 1:50)
data[["predictions_subset"]] <- predict_assay_subset

DefaultAssay(data) <- "predictions_subset"
```

```r
predict_assay_subset <- as.data.frame(data@assays$predictions_subset@data)
head(predict_assay_subset, 30)
```

```r
write.csv(predict_assay_subset, '0710_predict_assay_4cluster.csv')
```
